## Make a bigger dataset

```python
x = np.arange(-100,100,4)
x, x.shape
```

```python
y = x+10
y, y.shape
```

```python
SPLIT_SIZE = int(len(x)*0.8)

x_train, y_train = x[0:SPLIT_SIZE], y[0:SPLIT_SIZE]
x_test, y_test = x[SPLIT_SIZE:], y[SPLIT_SIZE:]

len(x_train), len(x_test)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(x_train, y_train, c='g', label="training_data")
plt.scatter(x_test, y_test, c='b', label="testing_data")
plt.legend()
```

```python
tf.random.set_seed(42)

# Create a model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=[1])
])

# Compile a model
model.compile(loss=["mae"],
              optimizer=tf.keras.optimizers.SGD(),
              metrics=["mae"])

model.summary()
```

```python
model.fit(x_train, y_train, epochs=100, verbose=1)
```

```python
from tensorflow.keras.utils import plot_model

plot_model(model, show_shapes=True)
```

```python
y_preds = model.predict(x_test)
y_preds
```

```python
plt.figure(figsize=(10,7))
plt.scatter(x_train, y_train, c='b', label="train_data")
plt.scatter(x_test, y_test, c='r', label="test_data")
plt.scatter(x_test, y_preds, c='g', label="predict")
plt.xlabel("x")
plt.xlabel("y")


plt.legend()
```

```python
def plot_result(x_train, y_train, x_test, y_test, prediction):
  plt.figure(figsize=(10,7))
  plt.scatter(x_train, y_train, c='b', label="train_data")
  plt.scatter(x_test, y_test, c='r', label="test_data")
  plt.scatter(x_test, prediction, c='g', label="predict")
  plt.legend()
```

```python
plot_result(x_train=x_train, y_train=y_train,
            x_test=x_test, y_test=y_test, prediction=y_preds)
```

```python
tf.metrics.mean_absolute_error(y_true=y_test,
                               y_pred=tf.reshape(y_preds, -1))
```

```python
tf.metrics.mean_absolute_error(y_true=y_test,
                               y_pred=y_preds.squeeze())
```

# Medical Cost Dataset

```python
import pandas as pd

insurance = pd.read_csv("https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/insurance.csv")

insurance.head()
```

```python
# Turn all categories into numbers
insurance_one_hot = pd.get_dummies(insurance)
insurance_one_hot.head()
```

```python
x = insurance_one_hot.drop("charges", axis=1)
y = insurance_one_hot["charges"]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=123)
```

## Model_0

```python
tf.random.set_seed(42)

model_0 = tf.keras.Sequential([
    tf.keras.layers.Dense(1),
    tf.keras.layers.Dense(1)
])

model_0.compile(loss=["mae"],
                optimizer=tf.keras.optimizers.SGD(),
                metrics=["mae"])

model_0.fit(x_train, y_train, epochs=100,
            callbacks=[tf.keras.callbacks.EarlyStopping(monitor="mae", 
                                                        patience=50, 
                                                        verbose=0,
                                                        restore_best_weights=True)])
```

```python
model_0_preds = model_0.predict(x_test)
model_0_preds[0:10]
```

```python
model_0.evaluate(x_test, y_test)
```

```python
tf.metrics.mean_absolute_error(tf.squeeze(model_0_preds), y_test)
```

## Model 1

```python
tf.random.set_seed(42)

model_1 = tf.keras.Sequential([
    tf.keras.layers.Dense(128),
    tf.keras.layers.Dense(64),
    tf.keras.layers.Dense(1)
])

model_1.compile(loss=["mae"],
                optimizer=tf.keras.optimizers.Adam(),
                metrics=["mae"])

history_1 = model_1.fit(x_train, y_train, epochs=100,
            callbacks=[tf.keras.callbacks.EarlyStopping(monitor="mae", 
                                                        patience=50, 
                                                        verbose=0,
                                                        restore_best_weights=True)])
```

```python
evaluate_model_1 = model_1.evaluate(x_test, y_test)
```

# Preprocessing Data

```python
insurance.head()
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder

ct = make_column_transformer(
    (MinMaxScaler(), ["age", "bmi", "children"]),
    (OneHotEncoder(handle_unknown="ignore"), ["sex","smoker","region"])
)

# Create x&y
x = insurance.drop("charges", axis=1)
y = insurance["charges"]

# Split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=42)

# Fit column transformer on the training data only
ct.fit(x_train)

# Transfor traingi and test data with normalization
x_train_normal = ct.transform(x_train)
x_test_normal = ct.transform(x_test)

```

```python
x_train_normal[0:5], x_train_normal.shape
```

## Model_2

```python
tf.random.set_seed(42)

model_2 = tf.keras.Sequential([
    tf.keras.layers.Dense(128),
    tf.keras.layers.Dense(32),
    tf.keras.layers.Dense(1)
])

model_2.compile(loss=["mae"],
                optimizer=tf.keras.optimizers.Adam(),
                metrics=["mae"])

history_2 = model_2.fit(x_train_normal, y_train, epochs=500,
            callbacks=[tf.keras.callbacks.EarlyStopping(monitor="mae", 
                                                        patience=100, 
                                                        verbose=0,
                                                        restore_best_weights=True)])
```

```python
evaluate_model_2 = model_2.evaluate(x_test_normal, y_test)
```
